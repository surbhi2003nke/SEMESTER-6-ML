**Data Visualization**

```python
plt.figure(figsize=(2,4))
sns.countplot(x = 'Outcome', data = data)
```

```python
col = data.columns
col
```

```python
plt.figure(figsize=(15, 15))
for idx, col_name in enumerate(col):
    plt.subplot(3, 3, idx + 1)
    sns.boxplot(y=col_name, data=data)

    plt.title(col_name, fontsize=15)

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)  # Added vertical spacing
plt.show()
```
